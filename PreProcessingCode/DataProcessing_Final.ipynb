```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')

#This is to clean up the text data and make it consistent
def refer_func(x):
    if 'search' in x:
        return 'search'
    elif 'message' in x:
        return 'message'
    elif 'tutorial' in x:
        return 'tutorialcomplete'
    elif 'course' in x:
        return 'coursecomplete'
    elif 'experiment' in x:
        return 'experimentStart'
    else:
        return x
    
df_data = pd.read_csv('geordi-1-5-17.csv',low_memory=False)
df_data.head(5)
```

```python
df_data['type'] = pd.Series(df_data['type'], dtype="string")
df_data['data'] = pd.Series(df_data['data'], dtype="string")


def replace_NA (x):
    if type(x) is not str:
        return ''
    else:
        return x
    
#Concatenating type and data
df_data['data'] = df_data['data'].apply(lambda x: replace_NA(x))
df_data['concat'] = df_data['type'] + ' ' + df_data['data']
df_data.head(10)
```

```python
df_data = df_data.iloc[:,[3,1,-1,9]]
df_data['event'] = df_data['concat'].apply(lambda x: refer_func(x))
df_data = df_data.iloc[:,[0,1,3,4]]
```

```python
#TramineR works best when we remove dash(-)
def removehdash(x):
    x = str(x)
    if '-' in x:
        return re.sub('-', '', x)
    else:
        return x
```

```python
df_data = df_data[df_data['userID'] != '(anonymous)']
df_data['event'] = df_data['event'].apply(lambda x: removehdash(x))
df_data['event'] = df_data['event'].astype('string')

df_data.reset_index(drop = True, inplace = True)
df_data['userID'] = pd.to_numeric(df_data['userID'], downcast = 'integer')
```

# Create Meta Labels

```python
df_meta_activity = pd.read_excel('MetaActivity.csv.xlsx')
#df_meta_activity = df_meta_activity.iloc[:,[1,3]]

def create_meta_labels(x):
    flag = False
    for event in range(len(df_meta_activity['Geordi_name'].values)):
        if df_meta_activity['Geordi_name'].values[event].strip() in x.strip():
            return df_meta_activity.iloc[event,1].strip()
    if (flag == False):
        return x
    
df_data['event'] = df_data['event'].apply(lambda x: create_meta_labels(x))
```

# Include Promotion timestamps for all users

```python
df_promotion = pd.read_csv('user_promotion.csv')
df_promotion['time'] = pd.to_datetime(df_promotion['Date'])
df_promotion = df_promotion.iloc[:,[1,2,4]]
def workflow_update(x):
    return "promotion_" + "level_" + x[0]
df_promotion['workflow_name'] = df_promotion['workflow_name'].apply(lambda x: workflow_update(x))
df_promotion.columns = ['userID','event', 'time']
df_promotion['userID'] = pd.to_numeric(df_promotion['userID'], downcast = 'integer')

df_data = pd.concat([df_data, df_promotion], ignore_index=True)

users = list(set(df_data['userID'].values))
```

```python
df_data = df_data[df_data['event'] != 'promotion_level_1']
df_data['sessionNumber'] = 0
```

# Assign Session numbers for each activity. Assign a new session number if inactivity time > 30 minutes

```python
for user in users:
    count = 1
    df_user = df_data[df_data['userID'] == user]
    df_user['time'] = pd.to_datetime(df_user['time']).values
    df_user.sort_values(by='time',inplace = True)
    df_user.reset_index(drop = False, inplace = True)
    for i in range(len(df_user)):
        if i == 0:
            df_user.iloc[i,3] = 1
            df_data.loc[df_user['index'].values[i], 'sessionNumber'] = 1
            continue
        else:
            cure_time = df_user.iloc[i,2]
            previ_time = df_user.iloc[i-1,2]
            timedelta = cure_time - previ_time
            if (timedelta.total_seconds())/60 > 30:
                count += 1
                df_user.iloc[i,3] = count
                df_data.loc[df_user['index'].values[i], 'sessionNumber'] = count
            else:
                df_user.iloc[i,3] = count
                df_data.loc[df_user['index'].values[i], 'sessionNumber'] = count
```

```python
session_nos = list(set(df_data['sessionNumber'].values))
```

# There are users with just promotion as their activity. No further activity available in the dataset. Let's remove such users

```python
users = list(set(df_data['userID'].values))
users_oneactivity = []
for user in users:
    df_user = df_data[df_data['userID'] == user]
    events = list(set(df_user['event'].values))
    events_promotion = [i for i in events if 'promotion' not in i]
    if 0 == len(events_promotion):
        users_oneactivity.append(user)
df_data = df_data[~df_data['userID'].isin(users_oneactivity)]
df_data.reset_index(drop = True, inplace = True)
```

```python
users = [user for user in users if user not in users_oneactivity]
```

# Store metrics for a given user, such as no. of sessions, session_no etc

```python
#This will store all session numbers for a user
users = list(set(df_data['userID'].values))
users_sessions = {}
for user in users:
    df_user = df_data[df_data['userID'] == user]
    sessions = list(set(df_user['sessionNumber'].values))
    users_sessions[user] = sessions
    
users_session_df = pd.DataFrame(columns= ['users', 'sessions'])
users_session_df['users'] = list(users_sessions.keys())
users_session_df['sessions'] = list(users_sessions.values())
users_session_df['no_sessions'] = users_session_df['sessions'].apply(lambda x: len(x))
users_session_df.to_csv("Users_Session_Info.csv")
```

```python
#This recursive function will count the number of consecutive classifies in a given series
def classify_counts(list_1,index, count):
    if index < len(list_1):
        if 'classify' in list_1[index]:
            return classify_counts(list_1, index+1, count + 1)
        else:
            return count
    else:
        return count
```

```python
users = list(set(df_data['userID'].values))
max_activity = 0
for user in users:
    df_temp_ = df_data[df_data['userID'] == user]
    if max_activity < len(df_temp_):
        max_activity = len(df_temp_)
max_sequence_length = max_activity + 1862
```

```python
df_sequence['users'] = users
df_sequence.set_index('users', inplace=True)
df_sequence.head()
```

```python
def classify_counts(list_1,index, count):
    if index < len(list_1):
        if 'classify' in list_1[index]:
            return classify_counts(list_1, index+1, count + 1)
        else:
            return count
    else:
        return count
```

```python
def reset_classify(x):
    if 'classify ' in x:
        return 'classify'
    elif ' classify' in x:
        return 'classify'
    else:
        return x
```

```python
def more_changes(activity):
    if (('logout' in activity) or ('login' in activity) or ('discussiontime' in activity) or ('changeprojectsidebar' in activity) or ('closefieldguide' in activity) or ('coursecomplete' in activity) or ('footermenu' in activity) or ('tutorialcomplete' in activity)):
        return 'remove'
    elif (('breadcrumb' in activity) or ('globemenu' in activity)):
        return 'exploring'
    elif ('message' in activity):
        return 'socialize'
    else:
        return activity
```

```python
df_data['event'] = df_data['event'].apply(lambda x: reset_classify(x))
df_data['event'] = df_data['event'].apply(lambda x: more_changes(x))
df_data = df_data[df_data['event'] != 'remove']
df_data['time'] = pd.to_datetime(df_data['time']).values
```

```python
for user in users:
    wanted_sessions = users_sessions[user]
    df_user = df_data[df_data['userID'] == user].copy()
    #df_user['time'] = pd.to_datetime(df_user['time']).values
    df_user = df_user.reset_index().sort_values(by = ["time","index"]).drop("index", axis = 1).reset_index(drop = True)
    events = []
    rows = 0
    for key in wanted_sessions:
        df_data_temp = df_user[df_user['sessionNumber'] == key].copy()
        #df_data_temp['time'] = pd.to_datetime(df_data_temp['time']).values
        #df_data_temp.sort_values(by = 'time', inplace = True)
        df_data_temp.reset_index(drop = False, inplace = True)
        len_df = df_data_temp.shape[0]
        first, last = False, False
        replace_first, replace_last, replace_mid = False, False, False
        if len(df_data_temp) != 0:
            length = -10
            break_point = 0
            skip_flag = False
            for i in range(len(df_data_temp['event'].values)):
                if (('classify' in df_data_temp['event'].values[i])  and (skip_flag == False)):
                    first, last = False, False
                    replace_first, replace_last, replace_mid = False, False, False
                    break_point = i
                    if (break_point == 0):
                        first = True
                    elif break_point == len_df-1:
                        last = True
                    length = classify_counts(df_data_temp['event'].values, i, 0)
                    previous_time_stamp, current_time_stamp, last_time_stamp1, last_time_stamp2 = 0,0,0,0
                    if (first != True and last != True):
                        previous_time_stamp = df_data_temp.iloc[i-1,2]
                        current_time_stamp = df_data_temp.iloc[i,2]
                        last_time_stamp1 = df_data_temp.iloc[i+length-1,2]
                        if (i + length == len_df):
                            last_time_stamp2 = last_time_stamp1
                        else:
                            last_time_stamp2 = df_data_temp.iloc[i+length,2]
                        if((current_time_stamp == previous_time_stamp) and (last_time_stamp1 == last_time_stamp2)):
                            replace_mid = True
                        elif((current_time_stamp != previous_time_stamp) and (last_time_stamp1 == last_time_stamp2)):
                            replace_first = True
                        elif((current_time_stamp == previous_time_stamp) and (last_time_stamp1 != last_time_stamp2)):
                            replace_last = True
                        elif((current_time_stamp != previous_time_stamp) and (last_time_stamp1 != last_time_stamp2)):
                            replace_first = True
                    elif (first == True):
                        if (i+length == len_df):
                            replace_last = True
                        else:
                            #print(i, length, df_data_temp.shape[0])
                            last_time_stamp1 = df_data_temp.iloc[i+length-1,2]
                            last_time_stamp2 = df_data_temp.iloc[i+length,2]
                            if (last_time_stamp1 == last_time_stamp2):
                                replace_first = True
                            else:
                                replace_last = True
                    skip_flag = True
                    #print('Status: i,replace_first, replace_last, replace_mid,length,skip_flag, first, last, previous_time_stamp, current_time_stamp, last_time_stamp1, last_time_stamp2 ', rows+i,replace_first, replace_last, replace_mid,length,skip_flag, first, last, previous_time_stamp, current_time_stamp, last_time_stamp1, last_time_stamp2)
                    if (length == 1 or replace_first == True):
                        #print("Updating first element: ")
                        if ((3 < length) and (length < 17)):
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_mid'
                            skip_flag = True

                        elif (length > 16):
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_high'
                            skip_flag = True
                        else:
                            if (length == 1):
                                df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_low'
                                skip_flag = False
                            else:
                                df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_low'
                                skip_flag = True
                    #print("Skip FLag after update first: ", skip_flag)
                    continue
                elif (skip_flag == True):
                    #print("i and skip flag in elif loop: ",rows+i, skip_flag)
                    if ( replace_mid == True and i < break_point+ int(length/2) -1):
                        continue
                    elif (replace_mid == True and i == break_point + int(length/2)):
                        if ((3 < length) and (length < 17)):
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_mid'
                        elif (length > 16):
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_high'
                        else:
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_low'
                        if (length == 2):
                            skip_flag = False
                        else:
                            skip_flag = True
                        continue
                    elif ( replace_last == True and i < break_point+length -1):
                        continue
                    elif (replace_last == True and i == break_point + length -1):
                        if ((3 < length) and (length < 17)):
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_mid'

                        elif (length > 16):
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_high'

                        else:
                            df_user.loc[df_data_temp['index'].values[i],'event'] = 'classify_low'
                        skip_flag = False
                        continue
                    elif (i == break_point + length -1):
                        #print("Value at Index row is ",df_data_temp['event'][i])
                        skip_flag = False
                        continue
            rows+=(i+1)
#     df_user['time'] = pd.to_datetime(df_user['time']).values
     #df_user = df_user.sort_values('time')
    df_user = df_user.reset_index().sort_values(by = "index")
    df_user = df_user[df_user['event'] != 'classify']
    df_user = df_user.reset_index(drop = True)
    events = []
    sessions_user = list(set(df_user['sessionNumber'].values))
    sessions_user = sorted(sessions_user, reverse = False)
    for session in wanted_sessions:
        df_user_session = df_user[df_user['sessionNumber'] == session]
#         df_user_session['time'] = pd.to_datetime(df_user_session['time']).values
#         df_user_session.sort_values(by='time',inplace = True)
        events = events +  list(df_user_session['event'].values) + ['EOS']
    if events != []:
        df_sequence.loc[user,1:len(events)] = events
```
