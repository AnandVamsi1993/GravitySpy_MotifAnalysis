```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')

#This is to clean up the text data and make it consistent
def refer_func(x):
    if 'search' in x:
        return 'search'
    elif 'message' in x:
        return 'message'
    elif 'tutorial' in x:
        return 'tutorialcomplete'
    elif 'course' in x:
        return 'coursecomplete'
    elif 'experiment' in x:
        return 'experimentStart'
    else:
        return x
    
df_data = pd.read_csv('geordi-1-5-17.csv',low_memory=False)
df_data.head(5)
```

```python
df_data['type'] = pd.Series(df_data['type'], dtype="string")
df_data['data'] = pd.Series(df_data['data'], dtype="string")


def replace_NA (x):
    if type(x) is not str:
        return ''
    else:
        return x
    
#Concatenating type and data
df_data['data'] = df_data['data'].apply(lambda x: replace_NA(x))
df_data['concat'] = df_data['type'] + ' ' + df_data['data']
df_data.head(10)
```

```python
df_data = df_data.iloc[:,[3,1,-1,9]]
df_data.head()
```

```python
df_data['event'] = df_data['concat'].apply(lambda x: refer_func(x))
df_data = df_data.iloc[:,[0,1,3,4]]
df_data.head(10)
```

```python
#TramineR works best when we remove dash(-)
def removehdash(x):
    x = str(x)
    if '-' in x:
        return re.sub('-', '', x)
    else:
        return x
```

```python
df_data = df_data[df_data['userID'] != '(anonymous)']
df_data['event'] = df_data['event'].apply(lambda x: removehdash(x))
df_data['event'] = df_data['event'].astype('string')
```

```python
df_data.reset_index(drop = True, inplace = True)
df_data['userID'] = pd.to_numeric(df_data['userID'], downcast = 'integer')
```

# Assign Session numbers for each activity. Assign a new session number if inactivity time > 30 minutes

```python
for user in users:
    count = 1
    df_user = df_data[df_data['userID'] == user]
    df_user['time'] = pd.to_datetime(df_user['time']).values
    df_user.sort_values(by='time',inplace = True)
    df_user.reset_index(drop = False, inplace = True)
    for i in range(len(df_user)):
        if i == 0:
            df_user.iloc[i,3] = 1
            df_data.loc[df_user['index'].values[i], 'sessionNumber'] = 1
            continue
        else:
            cure_time = df_user.iloc[i,2]
            previ_time = df_user.iloc[i-1,2]
            timedelta = cure_time - previ_time
            if (timedelta.total_seconds())/60 > 30:
                count += 1
                df_user.iloc[i,3] = count
                df_data.loc[df_user['index'].values[i], 'sessionNumber'] = count
            else:
                df_user.iloc[i,3] = count
                df_data.loc[df_user['index'].values[i], 'sessionNumber'] = count
```

```python
#This recursive function will count the number of consecutive classifies in a given series
def classify_counts(list_1,index, count):
    if index < len(list_1):
        if 'classify' in list_1[index]:
            return classify_counts(list_1, index+1, count + 1)
        else:
            return count
    else:
        return count
```

# This routine will check the length of classifies in the entire dataframe and encapsulate as 'mid-classify', low and high¶
#classify_mid : 4-16 #classify_high : >=17 #classify_low : <4

```python
for user in users:
    wanted_sessions = [1]
    df_data_temp = df_data[df_data['userID'] == user]
    #session_metrics_user = dict((k, session_metrics[k]) for k in wanted_sessions if k in session_metrics)
    for key in wanted_sessions:
        df_data_temp = df_data[(df_data['sessionNumber'] == key) & (df_data['userID'] == user)]
        df_data_temp['time'] = pd.to_datetime(df_data_temp['time']).values
        df_data_temp.sort_values(by = 'time', inplace = True)
        df_data_temp.reset_index(drop = False, inplace = True)
        if len(df_data_temp) != 0:
            length = -10
            break_point = 0
            skip_flag = False
            for i in range(len(df_data_temp['event'].values)):
                if (('classify' in df_data_temp['event'].values[i])  and (skip_flag == False)):
                    break_point = i
                    length = classify_counts(df_data_temp['event'].values, i, 0)
                    #print(length)
                    skip_flag = True
                    if (length == 1):
                        skip_flag = False
                        df_data.iloc[df_data_temp['index'].values[i],3] = 'classify_low'
                    #print(df_data_temp['index'].values[i])
                elif (skip_flag == True):
                    if ( i < break_point+length -1):
                        continue
                    elif ( i == break_point + length -1):
                        if (3 < length) & (length < 17):
                            df_data.iloc[df_data_temp['index'].values[i],3] = 'classify_mid'

                        elif (length > 16):
                            df_data.iloc[df_data_temp['index'].values[i],3] = 'classify_high'

                        else:
                            df_data.iloc[df_data_temp['index'].values[i],3] = 'classify_low'
                        skip_flag = False
                        continue
```

```python
#This recursive function will count the number of consecutive communals in a given series
def communal_counts(list_1,index, count):
    if index < len(list_1):
        if 'communal' in list_1[index]:
            return communal_counts(list_1, index+1, count + 1)
        else:
            return count
    else:
        return count
```

```python
def discussion_counts(list_1,index, count):
    if index < len(list_1):
        if 'view discussions' in list_1[index]:
            return discussion_counts(list_1, index+1, count + 1)
        else:
            return count
    else:
        return count
```

```python
for user in df_sequence.index.values:
    row_previous = list(df_sequence.loc[user,:].values)
    row_previous = [event for event in row_previous if str(event) != "0"]
    start = 1
    while (start < len(row_previous)+1):
        if str(df_sequence.loc[user, start]) == "communal":
            count = communal_counts(row_previous,start, 1)
            if count > 3:
                df_sequence.loc[user, start] = 'communal_high'
            elif(count > 0):
                df_sequence.loc[user, start] = 'communal_normal'
            start += count
        else:
            start+=1
```
